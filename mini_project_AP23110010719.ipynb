```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

tips['tip_pct'] = tips['tip'] / tips['total_bill']

piv = tips.pivot_table(
    index='day',
    columns='time',
    values='tip_pct',
    aggfunc='mean'
)

top_combos = piv.stack().sort_values(ascending=False).head(5)
print("Top 5 day-time combinations with highest average tip %:\n")
print(top_combos)

plt.figure(figsize=(6, 4))
plt.imshow(piv.values, aspect='auto', cmap='YlGnBu')
plt.title('Mean tip % by Day & Time')
plt.xlabel('Time of Day')
plt.ylabel('Day of Week')

plt.xticks(ticks=range(len(piv.columns)), labels=piv.columns.astype(str))
plt.yticks(ticks=range(len(piv.index)), labels=piv.index.astype(str))

plt.colorbar(label='Mean tip percentage')

plt.show()
```
